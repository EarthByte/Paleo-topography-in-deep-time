```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# =====================================================
# Function to annotate regression statistics
# =====================================================
def annotate_regression(ax, slope, intercept, r_squared, n):
    text = (
        f"N = {n}\n"
        f"y = {slope:.2f}x + {intercept:.2f}\n"
        f"$R^2$ = {r_squared:.3f}"
    )

    ax.text(
        0.05, 0.80,
        text,
        transform=ax.transAxes,
        fontsize=12,
        verticalalignment="top",
        bbox=dict(boxstyle="round", facecolor="white", alpha=0.8)
    )

# =====================================================
# File path
# =====================================================
csv_path = "/Users/jianpingzhou/work/Papers/Geology/Trainining dataset/gplate litho/Alfonso_Litho_Continental arc.csv"

# =====================================================
# Load CSV
# =====================================================
df = pd.read_csv(csv_path)

# Drop rows with missing values
df = df.dropna(subset=["Elevation_km", "Moho_km"])

# Sample number
n_samples = len(df)

# Extract variables
x = df["Moho_km"]
y = df["Elevation_km"]

# =====================================================
# Linear regression
# =====================================================
slope, intercept, r_value, p_value, std_err = linregress(x, y)
r_squared = r_value**2
regression_line = slope * x + intercept

# Isostatic balance point
balance_point_moho = -intercept / slope

# =====================================================
# Plot
# =====================================================
fig, ax = plt.subplots(figsize=(8,6))

ax.scatter(x, y, label="Data", alpha=0.7)
ax.plot(x, regression_line, color="red", label="Linear fit", linewidth=2)

# Add regression statistics with sample number
annotate_regression(ax, slope, intercept, r_squared, n_samples)

# Mark balance point
ax.scatter(balance_point_moho, 0, color="green", zorder=5,
           label=f"Balance point (Moho = {balance_point_moho:.2f} km)")

ax.axhline(0, color="black", linewidth=0.8, linestyle="--")

# Annotate balance point
balance_point_text = f"Balance point (Moho = {balance_point_moho:.2f} km)"
ax.text(
    0.95, 0.05,
    balance_point_text,
    transform=ax.transAxes,
    fontsize=12,
    verticalalignment="bottom",
    horizontalalignment="right",
    color="green",
    bbox=dict(boxstyle="round", facecolor="white", alpha=0.8)
)

# Labels
ax.set_xlabel("Moho depth (km)")
ax.set_ylabel("Elevation (km)")
ax.set_title("Linear Correlation between Moho Depth and Elevation (Continental Arc)")
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.savefig(
    "JP_Continent_arc_crust_elevation_correlation_1-1.pdf",
    format="pdf",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import linregress

# File path
csv_path = "/Users/jianpingzhou/work/Papers/Continous raster dem modeling/Trainining dataset/gplate litho/Alfonso_Litho_Island arc.csv"

# Load CSV
df = pd.read_csv(csv_path)

# Drop rows with missing values in the relevant columns
df = df.dropna(subset=["Elevation_km", "Moho_km"])

# Extract the variables
x = df["Moho_km"]
y = df["Elevation_km"]

# Compute linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)
r_squared = r_value**2
regression_line = slope * x + intercept

# Find the crustal thickness at 0 elevation (isostatic balance point)
balance_point_moho = -intercept / slope

# Create plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, label="Data", alpha=0.7)
plt.plot(x, regression_line, color="red", label="Linear fit", linewidth=2)

# Annotate equation and R² (moved down)
equation_text = f"y = {slope:.2f}x + {intercept:.2f}\n$R^2$ = {r_squared:.3f}"
plt.text(0.05, 0.80, equation_text, transform=plt.gca().transAxes,
         fontsize=12, verticalalignment='top', bbox=dict(boxstyle="round", facecolor="white", alpha=0.8))

# Mark the isostatic balance point (where elevation is 0)
plt.scatter(balance_point_moho, 0, color='green', zorder=5, label=f"Balance point (Moho = {balance_point_moho:.2f} km)")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')

# Annotate the balance point (bottom-right corner)
balance_point_text = f"Balance point (Moho = {balance_point_moho:.2f} km)"
plt.text(0.95, 0.05, balance_point_text, transform=plt.gca().transAxes,
         fontsize=12, verticalalignment='bottom', horizontalalignment='right', color='green', 
         bbox=dict(boxstyle="round", facecolor="white", alpha=0.8))

# Labels and formatting
plt.xlabel("Moho depth (km)")
plt.ylabel("Elevation (km)")
plt.title("Linear Correlation between Moho Depth and Elevation_Orogen")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig('JP_Continent arc crust-elevation correlation_1-3.pdf', format='pdf', dpi=300, bbox_inches='tight')
plt.show()
```
